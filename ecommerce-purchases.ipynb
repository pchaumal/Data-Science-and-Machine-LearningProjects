# Ecommerce Purchases Exercise

The dataset contains 14 columns and 10000 rows. Here's a summary of the columns:

1. Address: customer address
2. Lot: lot number of the property where customer's address is located
3. AM or PM: whether the purchase was made in the morning or evening
4. Browser Info: information about the customer's browser
5. Company: name of the company where the customer works
6. Credit Card: credit card number of the customer
7. CC Exp Date: expiration date of the customer's credit card
8. CC Security Code: security code of the customer's credit card
9. CC Provider: name of the company that issued the customer's credit card
10. Email: email address of the customer
11. Job: occupation of the customer
12. IP Address: IP address of the customer
13. Language: language preference of the customer
14. Purchase Price: price of the purchased item in USD (US Dollars)

The data types of the columns are:
- 11 columns are of object type (strings)
- 2 columns are of int64 type (integer)
- 1 column is of float64 type (float)

____
### Import pandas and read in the Ecommerce Purchases csv file and set it to a DataFrame called ecom. **

```python
# Importing necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Check the head of the DataFrame.**

```python
# Reading the csv file into a Pandas dataframe
ecom = pd.read_csv('/kaggle/input/ecommercepurchasesdata/EcommercePurchases.csv')
ecom.head()
```

** How many rows and columns are there? **

```python
ecom.info()
```

### What is the average Purchase Price? **

```python
# Calculate the average Purachase Price

average_purchase_price = ecom['Purchase Price'].mean()

# Print the result
print(f'The average Purachse Price is : {average_purchase_price}')
```

### What were the highest and lowest purchase prices? **

```python
# find the highest and lowest purchase prices

highest_price = ecom['Purchase Price'].max()
lowest_price = ecom['Purchase Price'].min()

# Print the results

print(f'The highest purchase price is ${highest_price:.2f}')
print(f'The lowest purchase price is ${lowest_price:.2f}')
```

### Histogram of the Purachse Price column

```python

# Set the style of the plot
plt.style.use('seaborn')

# Create a histogram of the Purachse Price column with 20 bins

plt.hist(ecom['Purchase Price'], bins=20, color='Green', edgecolor='white')

# Add labels to the plot
plt.xlabel('Purchase Price (USD)')
plt.ylabel('Number of Purchase')
plt.title('Distribution of Purchase Prices')

# Add grid lines to the plot
plt.grid(axis='y', alpha=0.5)

# Show the plot

plt.show
```

** How many people have English 'en' as their Language of choice on the website? **

```python
# Count the number of people who have chosen 'en as their Language of choice
num_en = (ecom['Language'] == 'en').sum()

# Print the result
print(f'{num_en} people have chosen English (en) as their Language of choice on the website.')
```

### How many people have the job title of "Lawyer" ? 


```python
# Count the number of people who have the job title of "Lawyer"
num_lawyers = (ecom['Job'] == 'Lawyer').sum()

# Print the result
print(f'{num_lawyers} people have the job title of "Lawyer".')
```

### How many people made the purchase during the AM and how many people made the purchase during PM ? **

**(Hint: Check out [value_counts()](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.value_counts.html) ) **

```python
# Count the number of people who made a purchase during the AM and PM
num_am_purchase = (ecom['AM or PM'] == 'AM').sum()
num_pm_purchase = (ecom['AM or PM'] == 'PM').sum()

# Print the results
print(f'{num_am_purchase} people made a purchase during AM\n{num_pm_purchase} people made a purchase during PM')
```

### What are the 5 most common Job Titles? **

```python
# Find the 5 most common jon titles
top_job_titles = ecom['Job'].value_counts().head(5)

# Print the result
print(f'The 5 most common job titles are:\n{top_job_titles}')
```

### Someone made a purchase that came from Lot: "90 WT" , what was the Purchase Price for this transaction? **

```python
# Filter the DataFrame to get the Purchase for the transactionthat came from LOT: '90 WT'

purchase_price = ecom[ecom['Lot'] == '90 WT']['Purchase Price'].values[0]

# Print the Purchase Price for the transaction that came from Lot :'90 WT'
print(f'The Purchase for the transaction that came from Lot: "90 WT" is ${purchase_price:.2f}')
```

### What is the email of the person with the following Credit Card Number: 4926535242672853 **

```python
# Find the row where Credit Card Number matches the given number

row = ecom.loc[ecom['Credit Card'] == 4926535242672853]

# Extract the Email value from the row
email = row['Email'].values[0]

# Print the email
print(f"The email of the person with the following Credit Card Number: 4926535242672853 is {email}")
```

### How many people have American Express as their Credit Card Provider *and* made a purchase above $95 ?**

```python
# Count the number of people who have American Express as their Credit Card Provider and made a purchase above $95
num_people = len(ecom[(ecom['CC Provider'] == 'American Express') & (ecom['Purchase Price'] > 95)])

# Print the result
print(f'{num_people} people have American Express as their Cridit Card Provider and made a purchase above $95.')
```

### How many people have a credit card that expires in 2025? **

```python
# Extract the year from the "CC Exp Date" column
ecom['Year'] = pd.to_datetime(ecom['CC Exp Date'], format='%m/%y').dt.year

# Count the number of people with a credit card that expires in 2025
count = (ecom['Year'] == 2025).sum()

# Print the result
print("Number of people with a credit card that expires in 2025:", count)

# Create a bar plot to visualize the result
plt.bar(['2025'], [count])
plt.title('Credit Cards Expiring in 2025')
plt.xlabel('Year')
plt.ylabel('Number of People')
plt.show()
```

### What are the top 5 most popular email providers/hosts (e.g. gmail.com, yahoo.com, etc...) **

```python
# Extract the domain from each email address and count their occurrences
domains = ecom['Email'].apply(lambda x: x.split('@')[1])
top_domains = domains.value_counts().head()

# Print the top 5 most popular domains
print('Top 5 most popular email providers/hosts:')
for domain, count in top_domains.iteritems():
    print(f'{domain}: {count}')
```

## Pie chart of payment methods:

```python
# Group the data by payment method and count the number of transactions
payment_counts = ecom.groupby('CC Provider')['Credit Card'].count()

# Create a pie chart
plt.figure(figsize=(8, 8))  # increase the size of the chart
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%')
plt.title('Payment Methods')
plt.show()
```

## Pie chart of purchase languages:

```python
# Group the data by language and count the number of transactions
lang_counts = ecom.groupby('Language')['Credit Card'].count()

# Create a pie chart
plt.figure(figsize=(8, 8))  # increase the size of the chart
plt.pie(lang_counts, labels=lang_counts.index, autopct='%1.1f%%')
plt.title('Purchase Languages')
plt.show()
```

## Pie chart of purchase prices:

```python
# Create bins for the purchase prices
bins = np.arange(0, 201, 20)

# Group the data by purchase price bin and count the number of transactions
price_counts = ecom.groupby(pd.cut(ecom['Purchase Price'], bins))['Credit Card'].count()

# Create a pie chart
plt.figure(figsize=(8, 8))  # increase the size of the chart
plt.pie(price_counts, labels=price_counts.index.astype(str), autopct='%1.1f%%')

# Move the labels outside the pie chart
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Purchase Prices')
plt.show()
```
